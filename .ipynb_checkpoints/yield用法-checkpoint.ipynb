1. 把yield看作 中断
2. 在碰到next()之后便会执行yield之后的语句 即继续

```python
def foo():
    print("starting...")
    while True:
        res = yield 7
        print('res:{}'.format(res))
        print("{:_>20}".format('_'))
        
g = foo() 
print(next(g)) # 这里会输出,'starting' 和 7，输出7之后便会中断
```

```python
print(next(g))# next()使进入函数时，从yield处开始，并在下一次yield处跳出
```

**至于为什么使用yield生成器：
节省空间**

```python
# 如我想要一个数列
# 我可用这样
a = [i for i in range(100)]
# 这样我得到一个序列，可是当序列较大时便会占用了不少内存

# 而有时，我只是要这个数列按顺序的值，那么我们就可用这样：
for i in range(100):
    pass
# 所以，其实range()就是生成器
```

```python
def fac(n):
    i, s = 1,1
    while n:
        s = s * i
        i += 1
        yield s
        n -= 1
```

```python
for i in fac(10):
    print(i)
```

```python
import numpy as np
data_x = np.linspace(0,10,100)
data_x = data_x.reshape(1,100)
data_y = np.sin(data_x)
data = np.concatenate((data_x,data_y),axis=0) 
```

```python
data.shape[1]
```

```python
# 写一个返回batchSize数据的shuffle的数据集：

def genData(data,batch_size):
    L = data.shape[1]
    idx = np.arange(L)
    np.random.shuffle(idx)
    num_batch = L//batch_size
    datax = data[0]
    datay = data[1]
    datax = datax[idx]
    datay = datay[idx]
    for i in range(num_batch):
        yield (datax[batch_size* i: batch_size*(i+1)], datay[batch_size* i: batch_size*(i+1)])
```

```python
for datax,datay in genData(data,10):
    print('data_x: {}\ndata_y: {}'.format(datax,datay))
    print('{:->90}'.format(''))
```
