```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xdata = np.linspace(0,10,100)
ydata = np.sin(xdata)
plt.plot(xdata,ydata)

x_value = np.arange(0,10,np.pi)
x_label = [0]
for i in range(len(x_value)):
    x_label.append('${}\\pi$'.format(i+1))

plt.xticks(x_value,x_label);
plt.yticks([]);
```

## 设置坐标轴比例，等

```python
theta = np.linspace(0,2*np.pi,100)
r = 5
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
plt.plot(x,y)
plt.axis('equal');
plt.axis('off');
```
