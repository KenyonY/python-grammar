### 创建一个二维表
pandas 通过**pd.DataFrame( data, index= , columns= )** 格式给出表的`index` 名称 `colums`名称以及数据data

```python
import pandas as pd
import numpy as np
data = np.ones([3,5]) # 3行 5 列
indexName = [1,2,3]
colName = [1,2,3,4,5]
pd.DataFrame(data,index = indexName,columns= colName)
```

### 通过字典创建
以及通过`.to_csv`写入csv文件

```python

a = [1,2,3]
b = [4,5,6]    

# ----------------------------------------------- 字典中的key值即为csv中列名
df0 = dataframe = pd.DataFrame({'a_name':a,
                          'b_name':b})

# ---------------------------------------------- 将DataFrame存储为csv,index表示是否显示行名，default=True
dataframe.to_csv("test.csv",index=True,sep=',')
```

### pandas读csv方法

```python
data = pd.read_csv('test.csv')
data
```

### 查看数据
以下是查看框架的顶行和底行的方法：

```python
data = np.random.randn(60, 40) #
indexName = np.linspace(0,59,60)
colName = np.linspace(1,40,40)
df = pd.DataFrame(data,index = indexName,columns= colName);

df.head()   # 查看最前5行

df.tail(10) # 查看最后10行
```

```python
df.describe() # 显示数据的快速统计摘要：
```

### `DataFrame.to_numpy()`
给出基础数据的NumPy表示。请注意，当您DataFrame拥有不同数据类型的列时，他的操作可能很昂贵，这可归结为pandas和NumPy之间的根本差异：NumPy数组对整个数组有一个dtype，而pandas DataFrames每列有一个dtype。当你访问 DataFrame.to_numpy()，pandas会找到可以容纳 DataFrame中所有 dtypes 的NumPy dtype。  
因为df，我们DataFrame的所有浮点值 DataFrame.to_numpy()都很快，不需要复制数据。

```python
df.to_numpy()
```
