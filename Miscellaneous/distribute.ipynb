```python
def win(x, delta=0.1):
    length = len(x)
    x = np.append(x, np.inf)
    res = np.zeros(length)
    
    for i in range(length):
        j = i+1
        cnt = 0
        while x[j] - x[i] <= delta:
            cnt += 1
            j += 1
        res[i] = cnt
    
    np.delete(x, -1)
    return res     
```

```python
plt.plot(win(norm,delta=0.5))
```
