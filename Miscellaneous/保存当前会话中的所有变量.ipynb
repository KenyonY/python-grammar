# 对于numpy中的npy 和 npz文件的存储与恢复

```python
# 对于单个数组的存储可以直接用np.save, 以.npy文件存储
a = np.arange(10)
np.save('a.npy', a)

# 恢复
a = np.load('a.npy')
a
```

```python
# 对于多个数组存放到一个文件中,使用np.savez(), 以.npz文件存储
b = np.arange(15)
np.savez('ab.npz', a=a, b=b) # 给变量命名为a, b 方便字典提取

# 恢复
AB = np.load('ab.npz')
a = AB['a']
b = AB['b']
print(a)
print(b)
```
