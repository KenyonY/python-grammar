```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xdata = np.linspace(0,10,100)
ydata = np.sin(xdata)
plt.plot(xdata,ydata)

x_value = np.arange(0,10,np.pi)
x_label = [0]
for i in range(len(x_value)):
    x_label.append('${}\\pi$'.format(i+1))

plt.xticks(x_value,x_label)
plt.yticks([])
```
